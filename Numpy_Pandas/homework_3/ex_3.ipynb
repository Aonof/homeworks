```python
import pandas as pd

movies = pd.read_csv('movies.csv')
movies.head()
```

```python
reviews = pd.read_csv('ratings.csv')
reviews.head()
```

```python
years = list(range(1950, 2017)) # в базе фильмы до 2016 года
print(years)
```

```python
import re

def production_year(row):
    ''' Возвращает год выпуска фильма '''
    pattern = r'\d{4}'
    year_search = re.findall(pattern, row)
    if year_search:
        year = int(year_search[-1]) # индекс -1 на случай, если есть год в названии фильма
        if year in years:
            return year
    return 1900
```

```python
# добавление столбца с годом выпуска фильма
movies['year'] = movies['title'].apply(production_year)
movies.head()
```

```python
# подсчет среднего рейтинга для каждого Id и преобразование в DataFrame
rating = reviews.groupby('movieId').apply(lambda data : data.rating.mean())
rating = pd.DataFrame({'movieId':rating.index, 'rating':rating.values})
rating.head()
```

```python
# объединение в один DataFrame
table = movies.merge(rating, how='inner') # параметр inner, чтобы в таблице не было фильмов без рейтинга
table.head()
```

```python
# подсчет среднего рейтинга для каждого года выпуска фильма и преобразование в DataFrame
year_rating = table.groupby('year').apply(lambda data : data.rating.mean())
year_rating = pd.DataFrame({'year':year_rating.index, 'rating':year_rating.values})
# сортировака по убыванию рейтинга
year_rating.sort_values(by='rating', ascending=False, inplace=True)
year_rating.head(30)
```
