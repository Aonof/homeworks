```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def search_document(number):
    """Возвращает индекс докумета в списке документов по номеру"""
    for item in documents:
        if number == item['number']:
            return documents.index(item)


def person():
    """Выводит владельца документа по номеру"""
    number = input('Введите номер документа: ')
    if search_document(number) == None:
        print('Документ не найден!')    
    else:
        print('Владелец документа:', documents[search_document(number)]['name'])
    return   


def search_directory(number):
    """Возвращает номер полки"""
    for key, value in directories.items():
        if number in value:
            return key
    

def directory():
    """Выводит номер полки, на которой хранится документ"""
    number = input('Введите номер документа: ')
    directory_number = search_directory(number)
    if directory_number == None:
        print('Документ не найден!')
    else:
        print('Документ находится на полке:', directory_number)
    return


def information():
    """Выводит список документов и подробную информацию по каждому из них"""
    for item in documents:
        print('Документ №:', item['number'], ', тип:', item['type'], ', владелец:', item['name'], 
              ', полка хранения:', search_directory(item['number']))
    return


def directory_list():
    """Возвращает список полок в виде строки"""
    dir_list = ''
    for key in directories.keys():
        dir_list += key + ', '
    return dir_list.strip(', ')


def del_directory():
    """Удаляет пустую полку"""
    number = input('Введите номер удаляемой полки: ')
    if number not in directories:
        print('Такой полки не существует. Текущий перечень полок:', directory_list())
    else:
        if directories[number] == []:
            del(directories[number])
            print('Полка удалена. Текущий перечень полок:', directory_list())
        else:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:', directory_list())
    return
            

def new_directory():
    """Создает новую полку"""
    number = input('Введите номер новой полки: ')
    if number in directories:
        print('Такая полка уже существует. Текущий перечень полок:', directory_list())
    else:
        directories[number] = []
        print('Полка добавлена. Текущий перечень полок:', directory_list())
    return


def new_document():
    """Создает новый документ, вносит его в список документов и нужную полку"""
    number = input('Введите номер документа: ')
    doc_type = input('Введите тип документа: ')
    person = input('Введите владельца документа: ')
    directory = input('Введите полку для хранения: ')
    if directory in directories:
        new_data = {'type' : doc_type, 'number' : number, 'name' : person}
        documents.append(new_data)
        directories[directory].append(number)
        print()
        print('Документ добавлен. Текущий список документов:', '\n')
        information()
    else:
        print()
        print('Такой полки не существует. Добавьте полку командой "as". Текущий список документов:', '\n')
        information()


def del_document():
    """Удаляет документ из списка и с полки"""
    number = input('Введите номер документа: ')
    print()
    index = search_document(number)
    key = search_directory(number)
    if index == None:
        print('Документ не найден в базе. Текущий список документов:', '\n')
        information()
    else:
        del(documents[index])
        directories[key].remove(number)
        print('Документ удален. Текущий список документов:', '\n')
        information()
    return    
        
        
def move_document():
    """Перемещает документ на другую полку"""
    number = input('Введите номер документа: ')
    directory = input('Введите номер полки: ')
    print()
    index = search_document(number)
    key = search_directory(number)
    if index == None:
        print('Документ не найден в базе. Текущий список документов:', '\n')
        information()
    elif directory not in directories:
        print('Такой полки не существует. Текущий перечень полок:', directory_list())
    else:
        directories[key].remove(number)
        directories[directory].append(number)
        print('Документ перемещен. Текущий список документов:', '\n')
        information()
        
        
def main():
    """Вызывает функции по вводимой команде"""
    command = ''
    while command != 'q':
        command = input('Введите команду: ')
        if command == 'p':
            person()
        elif command == 's':
            directory()
        elif command == 'l':
            information()
        elif command == 'ads':
            new_directory()
        elif command == 'ds':
            del_directory()
        elif command =='ad':
            new_document()
        elif command == 'd':
            del_document()
        elif command == 'm':
            move_document()
    return

            
main()
```
